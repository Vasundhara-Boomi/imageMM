```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load an example image in grayscale
image = cv2.imread('image.png', cv2.IMREAD_GRAYSCALE)
if image is None:
    print("Error: Could not load the image.")
else:
    # Display the original image
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    plt.show()
```

```python
# Perform contrast stretching
min_val, max_val = np.min(image), np.max(image)
contrast_stretched = cv2.normalize(image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)

# Display contrast-stretched image
plt.imshow(contrast_stretched, cmap='gray')
plt.title('Contrast-Stretched Image')
plt.axis('off')
plt.show()
```

```python
# Apply a simple linear filter (Gaussian Blur)
filtered_image = cv2.GaussianBlur(image, (5, 5), 0)

# Display filtered image
plt.imshow(filtered_image, cmap='gray')
plt.title('Filtered Image (Gaussian Blur)')
plt.axis('off')
plt.show()
```

```python
# Plot histograms of the original, contrast-stretched, and filtered images
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.hist(image.ravel(), bins=256, range=[0, 256])
plt.title('Histogram of Original Image')

plt.subplot(1, 3, 2)
plt.hist(contrast_stretched.ravel(), bins=256, range=[0, 256])
plt.title('Histogram of Contrast-Stretched Image')

plt.subplot(1, 3, 3)
plt.hist(filtered_image.ravel(), bins=256, range=[0, 256])
plt.title('Histogram of Filtered Image')
plt.show()
```
